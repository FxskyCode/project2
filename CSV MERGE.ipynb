```python
#importing
import pandas as pd
import numpy as np
import re
```

```python
# calling csv files
df = pd.read_csv("../project2/data/games_clean.csv", encoding= "unicode_escape")
df1 = pd.read_csv("../project2/data/elo_diff_win_percentage.csv", encoding= "unicode_escape")
```

```python
df1
```

```python
#function for closer value on elo range table
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

```python
array = df1["elo_diff"]
```

```python
#making new columns with the value of elo difference
elo_diff = []
for i in df["rating_diff"]:
    x = find_nearest(array, value=i)
    elo_diff.append(x)
```

```python
# making the new column
df["elo_diff"] = elo_diff
```

```python
df
```

```python
# funtion to extract array of the len of the data frame with the win percentage
win_percentage = []
for i in df["elo_diff"]:
    x = df1.loc[(df1["elo_diff"] == i)]
    for i in x.win_percentage:
        z = i
        win_percentage.append(z)        
```

```python
#adding the new column based on the array extracted before
df["expected_win_percentage"] = win_percentage
```

```python
#exporting final csv for visualization
df.to_csv("data/final.csv", index = False, encoding='unicode_escape')
```
