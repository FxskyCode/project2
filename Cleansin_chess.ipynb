## CLEANSING CSV

```python
#importing
import pandas as pd
import numpy as np
import re
```

```python
# calling csv
df = pd.read_csv("../project2/data/games.csv", encoding= "unicode_escape")
```

```python
# changing values to be able to see all the columns
pd.set_option('display.max_columns', None)
```

```python
#cheking all the null values
df.isna().sum()
```

```python
#looking for leftover columns
df.sample()
```

```python
castle_list = []
for i in df["moves"]:
    word = i.split()
    if "O-O" in word:
        castle_list.append("Yes")
    else:
        castle_list.append("No")

castle_list
```

```python
df["kings_castle"] = castle_list
```

```python
#Delete all colums that we will not use
df = df.drop(["id","rated","created_at","last_move_at","increment_code","white_id","black_id","moves",], axis = 1)
#shape check 
df.shape
```

```python
df
```

```python
#mean of average movements
df["turns"].mean()
```

```python
#checking values
df["victory_status"].value_counts()
```

```python
#looking for the average elo of both players 
df["white_rating"].mean()
```

```python
#opening eco values list
df["opening_eco"].value_counts()
```

```python

df["opening_ply"].value_counts()
```

```python
#looking for all the values in opening
df["opening_name"].value_counts()
```

```python
# Delete everythings its after the dots and dots included to make it simple
df["opening_name"] = df["opening_name"].str.replace("|",":")
df["opening_name"] = df["opening_name"].str.replace("#",":")
df["opening_name"] = df["opening_name"]+":"
df['opening'] = df['opening_name'].str.extract('^(.+?):')
```

```python
df
```

```python
#checking cleansing results
df["opening"].value_counts()
```

```python
#removing old column
df = df.drop(["opening_name"], axis = 1)
```

```python
df
```

```python
#making new colum elo difference between players
df['rating_diff'] = df['white_rating'] - df['black_rating']
df
```

```python
#Making new column base on expected winner using rate difference
df['expected_winner'] = ['white' if x >= 0 else 'black' for x in df['rating_diff']]
df
```

```python
df["kings_castle"].value_counts()
```

```python
#exporting clean csv 
df.to_csv("data/games_clean.csv", index = False, encoding='unicode_escape')
```
